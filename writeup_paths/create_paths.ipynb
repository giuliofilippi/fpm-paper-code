# Train Path 1A

```python
# Define the vertices of the triangle
vertices = np.array([[0,0],[180-30,0],[180-30,38],[180+130,38],[180+130,0],[360,0]])
# number of images
num_images = 30
```

```python
# list of training images
train_images = []
# loop over distances from center
for r in np.linspace(0,0.6,num_images):
    # get location
    L = get_coord(r, pw=0)
    # get rotations
    im =  get_single_rotations(vertices, L)
    # append
    train_images.append(im)
```

```python
plot_images(train_images[::10])
```

```python
FPM = compute_fpm(train_images[0])
FPM
```

```python
x_val = x_from_fpm(train_images[0], FPM)
x_val - 150
```

```python
# save and export train images vector
train_images_vector = np.array(train_images)
np.save('train/path_1a.npy', train_images_vector)
```
