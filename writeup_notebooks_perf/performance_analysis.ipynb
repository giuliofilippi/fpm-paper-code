# Prepare Data

```python
performance_tensor_ = np.load("performance_tensor.npy")
performance_vector_ = np.load("performance_vector.npy")
```

```python
# need to average over 10 runs
performance_tensor = np.mean(performance_tensor_, axis=0)
performance_vector = np.mean(performance_vector_, axis=0)
```

# Results Analysis Overall

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
performance_matrix = np.mean(performance_tensor, axis=0)
performance_matrix
```

```python
# Step 2: Assign labels to the rows and columns
row_labels = [0, 0.1, 0.2, 0.3, 0.4]
column_labels = [0, 8, 16, 24, 32, 40]

# Step 3: Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(performance_matrix, annot=True, fmt=".2f", cmap="viridis", xticklabels=column_labels, yticklabels=row_labels)
plt.title('Heatmap of MAE Values (Overall)')
plt.xlabel('Overlap')
plt.ylabel('Crossover')

# Save the plot as an SVG file
plt.savefig("../writeup_notebooks_perf/perf_exports/perf_overall.svg", format="svg", bbox_inches='tight')

plt.show()
```
